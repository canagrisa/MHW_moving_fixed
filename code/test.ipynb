```python
import xarray as xr
import numpy as np
```

```python
file = '/home/prossello/MHW_article/data/satellite/sst_rep_l4/sst_1982_2021.nc'

ds = xr.open_dataset(file)
```

```python
lat_min, lon_min = 38.437618, 0.825540
lat_max, lon_max = 39.526359, 1.929534

ds_c = ds.sel(lon=slice(lon_min, lon_max), lat=slice(lat_min, lat_max))
ds_c.analysed_sst.isel(time=0).plot()
```

```python
ds_c = ds_c.astype(np.float32)

ds_c.to_netcdf('/home/prossello/MHW_article/data/satellite/sst_1982_2021_ibiza.nc')
```
